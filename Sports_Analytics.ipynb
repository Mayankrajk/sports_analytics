```python
import ast

def extract_items(obj):
    data = ast.literal_eval(obj)
    selected_items = {
        'Full Name': data.get('Full Name'),
        'Born': data.get('Born'),
        'Age': data.get('Age'),
        'Batting Style': data.get('Batting Style'),
        'Bowling Style': data.get('Bowling Style'),
        'Playing Role': data.get('Playing Role'),
        'Fielding Position': data.get('Fielding Position'),
        'Teams': data.get('Teams'),
        'Also Known As': data.get('Also Known As')
    }
    return selected_items
```

```python
def convert_age(age_str):

    parts = age_str.split()


    if len(parts) > 0:

        years = int(parts[0].replace('y', ''))


        if len(parts) > 1:

            days = int(parts[1].replace('d', ''))

            years += days / 365


        return int(years)
    else:
        return None
```

```python
#Function to select Final region out of 3 to get correct country names
def new_co(df3):
    if df3['Region1'] is not None:
        if df3['Region1'].isupper():
            if (df3.Region1=='ICC'):
                return df3['Region2']
            else:
                return df3['Region1']

        elif df3['Region'] is not None:
            if df3['Region'].isupper():
                if df3['Region'] is not "ICC":
                    return df3['Region']
            else:
                return df3['Region2']
        else:
                return df3['Region2']
    else:
        return "NA"
```
